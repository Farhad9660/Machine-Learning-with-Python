## Training with all variables

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X.shape
```

```python
y = y.reshape(-1,1)
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.coef_
```

### R^2 Score

```python
R_2_score = reg.score(X_test, y_test)
```

```python
R_2_score
```

```python
RMSE = np.sqrt(mean_squared_error(y_pred, y_test))
```

### Root Mean Square Error

```python
RMSE
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
test_samples = np.arange(0, len(y_pred)).reshape(-1,1) 
#plotting the target values and predicted values
fig = plt.figure(figsize = (12,8))
sns.set_theme(palette = 'muted', font_scale = 1.5, context = 'paper')
plt.plot( test_samples , y_test, label = 'target values ', lw = 2)
plt.plot( test_samples , y_pred, label ='predicted values', lw = 2)
plt.plot( test_samples , (np.sqrt(np.square(y_pred - y_test))), label ='Square Error', lw = 2)
plt.xlabel('Test Samples')
plt.ylabel('MEDV(House Price. 1000 $)')
plt.legend(facecolor = 'white')
# plotting the regression line

plt.show()
```


## Regularized Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

### Ridge

```python
ridge = Ridge(alpha = 0.2, normalize = True )
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 43)
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_test, y_test)
```

### Lasso

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha = 0.2, normalize = True )
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 43)
```

```python
lasso.fit(X_train, y_train)
lasso.score(X_test, y_test)
```

```python
lasso.coef_
```

```python
coefs = lasso.fit(X_train, y_train).coef_
```

```python
coefs 
```

```python
boston.feature_names
```

#### Plotting the coefficients

```python
fig = plt.figure(figsize = (10,6))
plt.style.use('fivethirtyeight')
plt.plot(range(len(boston.feature_names)) , coefs, label = 'Lasso Coefficient for Boston Housing Data ', lw = 2)
plt.xticks(range(len(boston.feature_names)),boston.feature_names, rotation = 60 )
plt.xlabel('Features')
plt.ylabel('Coefficents')
plt.legend(facecolor = 'white')
# plotting the regression line

plt.show()
```
